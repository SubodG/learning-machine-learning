```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('heart.csv')
```

Changing sex to is_male 

```python

data['Sex'] = data['Sex'].apply(lambda x: 1 if x == 'male' else 0)
data.rename(columns={'Sex':'is_male'},inplace=True)
```

```python
data['ChestPainType'].value_counts()
```

```python
data = pd.get_dummies(data,columns=['ChestPainType'])
```

```python
data['ChestPainType_ASY'] = data['ChestPainType_ASY'].astype(int)
data['ChestPainType_ATA'] = data['ChestPainType_ATA'].astype(int)
data['ChestPainType_NAP'] = data['ChestPainType_NAP'].astype(int)
data['ChestPainType_TA'] = data['ChestPainType_TA'].astype(int)


```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['RestingECG'] = le.fit_transform(data['RestingECG'])
```

```python
data['ExerciseAngina'] = data['ExerciseAngina'].apply(lambda x: 0 if x == 'N' else 1)
```

```python
data['ST_Slope'] = data['ST_Slope'].apply(lambda x: 0 if x == 'Flat' else(1 if x == 'Up' else -1))
```

```python
x = data.drop('HeartDisease', axis=1)
y = data['HeartDisease']
```

```python
ros = RandomOverSampler(random_state=42)
x_resampled, y_resampled = ros.fit_resample(x, y)
```

```python
data = x_resampled.join(y_resampled)
```

```python
x = data.drop('HeartDisease', axis=1)
y = data['HeartDisease']
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc_model = RandomForestClassifier()
rfc_model.fit(x_train,y_train)
```

```python
rfc_model.score(x_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(x_train,y_train)
```

```python
lr_model.score(x_test,y_test)
```

```python
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(x_train,y_train)
svc_model.score(x_test,y_test)
```

```python
import pickle

# Save the RandomForestClassifier model to a file
with open('rfc_model.pkl', 'wb') as file:
    pickle.dump(rfc_model, file)
```

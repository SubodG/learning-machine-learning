```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
columns = ['sepal-length','sepal-width','petal-length','petal-width','class']
```

```python
data = pd.read_csv('iris.data',names = columns)
```

```python
data['class'].value_counts()
```

```python
data['class'] = data['class'].apply(lambda x: 0 if x == 'Iris-setosa' else(1 if x == 'Iris-versicolor' else 2))
```

```python
data.hist(figsize = (15,8))
```

```python
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc_model = RandomForestClassifier()
```

```python
x = data.drop(columns = 'class')
y = data['class']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
rfc_model.fit(x_train,y_train)
```

```python
rfc_model.score(x_test,y_test)
```

```python
#output_function
def predict_flower_class(sl,sw,pl,pw):
    input = [[sl,sw,pl,pw]]
    y = rfc_model.predict(input)
    output = "Isis-setosa"  if y == 0 else ("Iris-versicolor" if y == 1 else 'Iris-virginica ') 
    return output
```

```python
predict_flower_class(6.4,3.1,5.5,1.8)
```

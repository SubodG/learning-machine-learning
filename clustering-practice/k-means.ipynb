```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets as make_blobs
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs

x, y = make_blobs(n_samples=1000, n_features=2, centers=3, random_state=23)
```

```python
plt.scatter(x[:,0],x[:,1],c = y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.cluster import KMeans
```

```python
#elbow method
wccs = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(x_train)
    wccs.append(kmeans.inertia_)

     
```

```python
plt.plot(range(1,11),wccs)
plt.xticks(range(1,11))
plt.xlabel('No of clusters')
plt.ylabel('wccs')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3,init='k-means++')
y_labels = kmeans.fit_predict(x_train)
plt.scatter(x_train[:,0],x_train[:,1],c = y_labels)
plt.show(
```

```python
y_pred_label = kmeans.predict(x_test)
```

```python
plt.scatter(x_test[:,0],x_test[:,1],c = y_pred_label)
plt.show()
```
